```python
# Code
#Self-employed
md['self_employeed'] = md['self_employeed'].astype(bool)

print(md['self_employeed'].value_counts())

#company size I think is fine as is
print(md['company_size'].value_counts())

#tech_company
vals = {'1.0':True,
        '0.0': False,
        'True': True,
        'False': False}
md['tech_company'] = md['tech_company'].replace(vals)
print(md['tech_company'].value_counts())

#tech_prole
md['tech_prole'] = md['tech_prole'].replace(vals)

print(md['tech_prole'].value_counts())

#mh_benefits
md['mh_benefits'] = md['mh_benefits'].replace('Not eligible for coverage / N/A', 'Not eligible for coverage / NA')
print(md['mh_benefits'].value_counts())
```
